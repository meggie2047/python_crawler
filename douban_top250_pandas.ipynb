```python
# 重置索引
df = df.reindex(range(1,251))
df.info()
```

```python
df["year"] = df["year"].apply(lambda x: x.replace("(", "").replace(")", "") if isinstance(x, str) else x)
df.head(5)
```

```python
df["year"].value_counts()

#year counts
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
year_top_10 = df['year'].value_counts().head(10)
plt.title = "douban250_year_counts"
plt.xlabel = "year"
plt.ylabel = "year_counts"
bars = plt.bar(year_top_10.index, year_top_10.values)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')

plt.show()
```

```python
top10_year_list = df["year"].value_counts().head(10).index
df_top10_year = df[df["year"].isin(top10_year_list)]
movies = df_top10_year.query('year == "2010"') #《让子弹飞》跟《黑天鹅》竟然是同一年的
movies
```

```python
actor_lst = []
for i in df["cast"]:
    if isinstance(i, str):
        actors = [actor.strip() for actor in i.split("/")]
        actor_lst.extend(actors)
print(len(actor_lst))
from collections import Counter
actors_count = Counter(actor_lst)
actors_count = sorted(actors_count.items(), key=lambda x: x[1], reverse=True)
actors_count #top1是张国荣，今天正好是4月1日; 榜单前十被<哈利波特>演员霸占着
```

```python
genre_lst = []
for i in df["type"]:
    if isinstance(i, str):
        genres = [genre.strip() for genre in i.split("/")]
        genre_lst.extend(genres)
print(len(genre_lst))
genre_counter = Counter(genre_lst)
genre_counter = sorted(genre_counter.items(), key=lambda x: x[1], reverse=True)
genre_counter #最受欢迎的题材组合式：剧情/爱情
```

```python
free_lst = []
vip_free_lst = []
free_type_lst = []
for i in df["websites"]:
    if isinstance(i, str):
        free_type = [website.split("(")[1].strip(")") for website in i.split(",") if "(" in website]
        free_type_lst.extend(free_type)
free_type_lst = set(free_type_lst)
free_type_lst
for i in df["websites"]:
    if isinstance(i, str):
        for website in i.split(","):
            site = website.split("(")[0].strip()
            if "VIP免费观看" in website or "单片付费" in website:
                vip_free_lst.append(site)
            elif "(免费观看)" in website:
                free_lst.append(site)
vip_free_lst = Counter(vip_free_lst)
free_lst = Counter(free_lst)
free_lst
vip_free_lst = sorted(vip_free_lst.items(), key=lambda x: x[1], reverse=True)
free_lst = sorted(free_lst.items(), key=lambda x: x[1], reverse=True)
vip_free_lst  #由于豆瓣标识不完整，这个统计不完整，很多vip免费观看的电影后面没有附上免费信息
```
